```python
from PIL import Image
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
filepath = "C:/Users/hamsa/Documents/UoD/Course work/SEM 1/Programming Languages/PY3/GrowLocations.csv"
df = pd.read_csv(filepath)
df.head()
```

```python
Using loc methdd slice/select only the latitude and longitude columns from the dataframe
use the rename function to interchange the latitude and longitude column names
filter the rows of the dataframe for the specified ranges if latitude and longitude values
print first few rows of the filtered dataframe
```

```python
df1 = df.loc[:, ["Latitude","Longitude"]]
df1.rename({'Latitude': 'Longitude','Longitude': 'Latitude'}, axis = "columns", inplace = True)
df1 = df1[(df1['Latitude'] > 50.681) & (df1['Latitude'] < 57.985) & (df1['Longitude'] > -10.592) & (df1['Longitude'] < 1.6848)]
df1.head()
```

```python
Using Matplotlib create a scatter plot of the data in df1 with 'Longitude' column as the x-axis and 'Latitude' column as the 
y-axis, and each point represented by a marker (s=25 sets markers size to 25).
Using PIL library open the image specified by Imagepath, convert it to a NumPy array, and assign to the variable img.
Overlay the scatter plot with the image using the imshow function from Matplotlib.
extent parameter specifies the extent of the image in data coordinates (longitude and latitude ranges). 
alpha=0.5 parameter sets the transparency of the image, allowing the scatter plot to become visible.
plt.show(): displays the combined scatter plot and image.
```

```python
df1.plot(x='Longitude', y='Latitude', kind='scatter', s=25)
Imagepath = "C:/Users/hamsa/Documents/UoD/Course work/SEM 1/Programming Languages/PY3/map7.png"
img = np.asarray(Image.open(Imagepath))
imgplot = plt.imshow(img, extent=(-10.592, 1.6848, 50.681, 57.985), alpha=0.5)
plt.show()
```
